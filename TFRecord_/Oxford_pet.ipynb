```python
# 데이터 로드 - 경로 설정
data_path = os.path.join("images","*")   # *는 glob에서 쓰는데 정규표현식("전체"를 의미한다)

data_list = glob(data_path)

len(data_list)
```

```python
data_list[0]  # 데이터 경로
```

```python
# 확장자 확인 jpg
file_extension = []
for data in data_list:
    basename = os.path.basename(data)  # 중요
    _, file  = basename.split('.')     # 중요
    
    if file in file_extension:   # unique한 확장자만 모아서 file_extension 리스트에 담아줌
        continue
    else :
        file_extension.append(file)
```

```python
file_extension
```

```python
# mat은 뭔지 모르니까 지워야지

for i,data in enumerate(data_list):  # 인덱스 준 이유
    basename = os.path.basename(data)
    _, file  = basename.split('.') 
    
    if file == 'mat':
        del data_list[i]   # 여기서 쓰려고!
```

```python
# mat 3개가 줄었네
len(data_list)
```

```python
# channel 수 확인
# 3채널이 있을 수 있고 흑백 1채널, 4채널도 있을 수 있음
# 방법 1) 4채널 -> 3채널로 바꾸거나
# 방법 2) 4채널을 지우자

# 방법2 - RGB만 이용할게
for i,data in enumerate(data_list):
    image = Image.open(data) # 중요
    mode = image.mode        # 중요
    
    # mode : 이미지의 채널 형태 (어떻게 생겼는지 알려준다)
    # 3채널이면 mode가 RGB
    if mode != 'RGB':
        del data_list[i]
```

```python
# 방법1
class_list = []
for data in data_list:
    basename = os.path.basename(data)
    label = os.path.splitext(basename)[0]
    
    label = re.sub("_\d+","",label)
    
    if label in class_list:
        continue
    else:
        class_list.append(label)   # 라벨의 갯수
class_list
```
